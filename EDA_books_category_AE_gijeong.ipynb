```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

class AutoEncoder(nn.Module):
    def __init__(self, input_size, hidden_sizes, dropout=0.1):
        super(AutoEncoder,self).__init__()
        self.e_layers = []
        self.d_layers = []
        decoder_out_size = input_size
        
        for hidden in hidden_sizes:
            self.e_layers.append(torch.nn.Linear(input_size, hidden))
            self.e_layers.append(torch.nn.ReLU())
            self.e_layers.append(torch.nn.Dropout(p=dropout))
            input_size = hidden
        
        hidden_sizes = list(reversed(hidden_sizes))
        input_size = hidden_sizes[0]
        hidden_sizes.append(decoder_out_size)

        for hidden in hidden_sizes[1:]:
            self.d_layers.append(torch.nn.Linear(input_size, hidden))
            self.d_layers.append(torch.nn.ReLU())
            self.d_layers.append(torch.nn.Dropout(p=dropout))
            input_size = hidden   
        


        self.encoder = nn.Sequential(*self.e_layers)
        self.decoder = nn.Sequential(*self.d_layers)
        self._init_weights()
    
    def _init_weights(self):
        for module_ in self.modules():
            if isinstance(module_, nn.Linear):
                torch.nn.init.xavier_normal_(module_.weight)
    
    def forward(self,x):
        # print(x)
        x = self.encoder(x)
        # x = torch.sigmoid(x)
        x = self.decoder(x)
        # x = torch.sigmoid(x)
        return x
    

    
```

```python
class AutoEncoder_loss_fn(nn.Module):
    """
    AutoRec_loss_fn
    
    Args:
        - loss_fn: (nn.Module) 사용할 Loss Function
    Shape:
        - Input1: (torch.Tensor) Model의 예측 결과. Shape: (batch size,)
        - Input2: (torch.Tensor) 정답. Shape: (batch size,)
        - Output: (torch.Tensor) Observable한 데이터에 대해서만 계산한 Loss. Shape: ()
    """
    def __init__(self, loss_fn):
        super(AutoEncoder_loss_fn, self).__init__()
        self.loss_fn = loss_fn
    
    def forward(self, pred, y):
        y_for_compute = y.clone().to('cpu')
        index = np.where(y_for_compute != 0) # FILL HERE : USE np.where & y_for_compute. WARNING: y를 사용 시, y의 device가 gpu일 경우 오류 발생 #
        loss = self.loss_fn(pred[index], y[index])
        return loss
```

```python
# hyperparameter

input_size = len(train.columns)
hidden_sizes = [2048,1024,256]
lr = 0.001
epochs = 100
batch_size = 512
print(input_size)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = AutoEncoder(input_size=input_size,hidden_sizes=hidden_sizes).to(device)
fn_loss = AutoEncoder_loss_fn(nn.MSELoss().to(device)).to(device)
optimizer = optim.Adam(model.parameters(model.parameters()), lr=lr, amsgrad=True)
```

```python
train_dataset = TensorDataset(torch.FloatTensor(np.array(train)))
valid_dataset = TensorDataset(torch.FloatTensor(np.array(valid)), torch.FloatTensor(np.array(valid_nan)))
test_dataset = TensorDataset(torch.FloatTensor(np.array(test)), torch.FloatTensor(np.array(test)))

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=1, shuffle=False )
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

```python
import os
import time


def make_dir(path):
        '''
        [description]
        경로가 존재하지 않을 경우 해당 경로를 생성하며, 존재할 경우 pass를 하는 함수입니다.

        [arguments]
        path : 경로

        [return]
        path : 경로
        '''
        
        if not os.path.exists(path):
            os.makedirs(path)
        else:
            pass
        return path
```

```python
import datetime
import time
now = time.localtime()
now_date = time.strftime('%Y%m%d', now)
now_hour = time.strftime('%X', now)
save_time = now_date + '_' + now_hour.replace(':', '')


save_path = make_dir(f'./{save_time}/')

for epoch in range(epochs):
    running_loss = 0.0
    for data in train_loader:
        # print(data[0].size())
        data = data[0].to(device)
        optimizer.zero_grad()
        # data = torch.stack(data).to(device)
        # data = data.to(device)
        recon = model(data)
        loss_ = fn_loss(recon, data)
        loss_.backward()
        optimizer.step()
        running_loss += loss_.item()
    if (epoch < 100 and epoch % 10 == 9) or epoch % 100 == 99:
        print('[Epoch %d] loss: %.10f' % (epoch + 1, running_loss / len(train_loader)))
    if epoch % 10 == 9:
        torch.save(model, save_path+f'epoch_{epoch+1}.pth')
```
